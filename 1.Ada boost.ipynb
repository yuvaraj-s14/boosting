```python
import pandas as pd
dataset=pd.read_csv("50_startups.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
regressor=AdaBoostRegressor(random_state=0,n_estimators=100)
regressor.fit(x_train,y_train)
```

```python
weight=regressor.estimator_weights_
```

```python
weight
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_adaboost.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open("finalized_model_adaboost.sav",'rb'))
result=loaded_model.predict([[1234,342,4565,1,0]])
result
```
